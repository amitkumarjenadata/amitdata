```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv('mnc_salary.csv',parse_dates=['DOJ'])
ds
```

```python
ds.dtypes
```

```python
ds=pd.DataFrame(data=ds)
```

```python
ds.drop(['Name','Weight'],axis=1,inplace=True)
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
import numpy as np
ds['Salary'].replace(np.NAN,ds['Salary'].mean,inplace=True)
ds['College'].replace(np.NAN,ds['College'].mean,inplace=True)
```

```python
ds.isnull().sum()
```
